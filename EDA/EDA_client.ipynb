```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_client = pd.read_csv('../../estsoft/data/client.csv')
```

```python
# categorical features
features_cat = ['county','is_business','product_type']

# plot distributions
for f in features_cat:
    plt.figure(figsize=(6,3))
    df_client[f].value_counts().sort_index().plot(kind='bar')
    plt.title('client data - ' + f)
    plt.grid()
    plt.show()
```

```python
# numerical features
# eic_count : 유럽 식별자 코드
# installed_capacity : 설치된 전지용량
features_num = ['eic_count','installed_capacity']

# plot distributions
for f in features_num:
    plt.figure(figsize=(6,3))
    df_client[f].plot(kind='hist', bins=100)
    plt.title('client data - ' + f)
    plt.grid()
    plt.show()
```
