```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# config
default_color_1 = 'darkblue'
default_color_2 = 'darkgreen'
default_color_3 = 'darkred'

pd.set_option('display.max_columns', None)
```

```python
# weather forecast data
df_weather_fc = pd.read_csv('./forecast_weather.csv')
df_weather_fc.info(show_counts=True)
```

```python
# preview
df_weather_fc.head(10)
```

```python
# plot temperature
plt.figure(figsize=(6,3))
df_weather_fc.temperature.plot(kind='hist', bins=100,
                               color=default_color_1)
plt.title('Temperature')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 3))
df_weather_fc.dewpoint.plot(kind='hist', bins=100, color='blue')
plt.title('dewpoint')
plt.grid()
plt.show()
```

```python
df_ex = df_weather_fc[(df_weather_fc.latitude==57.6) & (df_weather_fc.longitude==21.7)]
```

```python
# plot time series for temperature
fig, ax = plt.subplots(figsize=(18,6))
ax.plot(df_ex.origin_datetime, df_ex.temperature, label='temperature',
        color=default_color_1)
ax.plot(df_ex.origin_datetime, df_ex.dewpoint, label='dewpoint',
        color=default_color_3)
ax.xaxis.set_major_locator(plt.MaxNLocator(20))
ax.legend(loc='upper left')
plt.xticks(rotation=90)
plt.title('Development of Temperature Forecast')
plt.grid()
plt.show()
```
